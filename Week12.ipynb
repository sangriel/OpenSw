```python
import csv

f = open("ta_20220519192644.csv","r",encoding="cp949")

data = csv.reader(f,delimiter=",")


maxTemp = -111111;
maxTempDate = ""
minTemp = 999999;
minTempDate = "";
maxDiff = 0;
maxDiffDate = "";


#날짜 , 지점, 평균,최저, 최고
for row in data:
    if len(row) == 5:
        if row[4] != "" and float(row[4]) > maxTemp:
            maxTemp = float(row[4])
            maxTempDate = row[0]

        if row[3] != "" and float(row[3]) < minTemp:
            minTemp = float(row[3])
            minTempDate = row[0]

        if row[3] != "" and row[4] != "" and  abs(float(row[3]) - float(row[4])) > maxDiff:
            maxDiff = abs(float(row[3]) - float(row[4]))
            maxDiffDate = row[0]
        
        
print("일교차가 가장 큰날")
print(maxDiffDate)
print("최저기온이 가장 낮은 날")
print(minTempDate)
print("최고기온이 가장 높은 날")
print(maxTempDate)
    
f.close()
```

```python
#사용월,호선명, 역Id, 지하철 역, 승차 승객수, 하차 승객수, 작업일시 
import csv
from IPython.lib.pretty import pretty


f = open("subway.csv","r")

csvFile = csv.reader(f,delimiter=",")

data = list(csvFile)

maxPassenger = {}
minPassenger = {}

def replaceMax(lineName,data):
    compareDict = maxPassenger[lineName]
    insertData = {"name": data[3], "value" : int(data[4].replace(",",""))}
    if len(compareDict) < 3:
        maxPassenger[lineName].insert(len(compareDict) - 1,insertData)
    else:
        for i in range(0,3):
            if compareDict[i]["value"] < int(data[4].replace(",","")):
                maxPassenger[lineName].insert(i,insertData)
                maxPassenger[lineName].pop()
                break
                
def replaceMin(lineName,data):
    compareDict = minPassenger[lineName]
    insertData = {"name": data[3], "value" : int(data[5].replace(",",""))}
    if len(compareDict) < 3:
        minPassenger[lineName].insert(len(compareDict) - 1,insertData)
    else:
        for i in range(0,3):
            if compareDict[i]["value"] > int(data[5].replace(",","")):
                minPassenger[lineName].insert(i,insertData)
                minPassenger[lineName].pop()
                break
    

for i in range(0,len(data)):
    if i == 0:
        maxPassenger[data[i][1]] = [{"name": data[i][3], "value" : int(data[i][4].replace(",",""))}]
        minPassenger[data[i][1]] = [{"name": data[i][3], "value" : int(data[i][5].replace(",",""))}]
    else:
        if len(data[i]) == 7:
            if data[i - 1][1] == data[i][1]:
                replaceMax( data[i][1],data[i])
                replaceMin( data[i][1],data[i])
            else:
                maxPassenger[data[i][1]] = [{"name": data[i][3], "value" : int(data[i][4].replace(",",""))}]
                minPassenger[data[i][1]] = [{"name": data[i][3], "value" : int(data[i][5].replace(",",""))}]
        




    
# print(pretty(maxPassenger))
# print(pretty(minPassenger))

for key,value in maxPassenger.items():
    print(key + " 이용객 가장 많은 역 순위")
    for i in range(0,3):
        print('    {0}. {1} ({2})'.format(i + 1,value[i]["name"], value[i]["value"]))
    print(key + " 이용객 가장 적은 역 순위")
    for i in range(0,3):
        print('    {0}. {1} ({2})'.format(i + 1,minPassenger[key][i]["name"],minPassenger[key][i]["value"]))
    
    print("---------------------------")

f.close()


```
